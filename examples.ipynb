Two examples in which :
* we generate a data set from a manifold
* we estimate the dimension

# Worms

```python
cloud = generators.implicit_sample(
        thickness=0.01,
        maxima=[20,14],
        n=500,
        dimension=2,
        function=generators.perlin_2d(
            grid_size=2,
            x_max=10,
            y_max=7,
            seed=25))
```

```python
d1 = estimators.corsum(cloud, e1=0.5*1.5, e2=0.5*0.19)
```

```python

plt.scatter(cloud[:,0],cloud[:,1])
plt.show()
```

# Schwarz surface

```python
cloud2 = generators.implicit_sample(
    n=2000, 
    dimension=3, 
    maxima=[4,4,4], 
    thickness=0.02, 
    function=generators.schwarz)
```

```python
d2 = estimators.corsum(cloud2, e1=0.78, e2=0.2, torus=True, maxima=[4,4,4])
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.scatter(cloud2[:,0],cloud2[:,1],cloud2[:,2], c=(cloud2[:,2]))
plt.show()
```
