```python
import scipy.integrate as integrate
from scipy.special import gamma
from math import pi, sqrt, sin, sinh, asin, log, exp, acos
import numpy as np
import random
import matplotlib.pyplot as plt
import pandas as pd
```

## Number of pairs for mean distance

```python
def mean(d):
    return d / (d+1)
def var(d):
    return d / ((d+2)*(d+1)**2)

result=[]
for d in range(2,11):
    gap = min( mean(d+1)-mean(d) , mean(d)-mean(d-1) )/2
    stdev = sqrt(var(d))
    n=pow(stdev/gap,2)*10
    result.append([d, gap, stdev, int(n)])
    
pd.DataFrame(result, columns=["dimension", "gap", "standard deviation", "nb of pairs"])
```

## Number of pairs for Corsum

```python
def mean(d):
    return (0.23/0.53)**d
def var(d):
    return mean(d)-mean(d)**2

result=[]
for d in range(2,11):
    gap = min( mean(d)-mean(d+1) , mean(d-1)-mean(d) )/2
    stdev = sqrt(var(d))
    n=pow(stdev/gap,2)*10
    result.append([d, gap, stdev, int(n)])
    
pd.DataFrame(result, columns=["dimension", "gap", "standard deviation", "nb of pairs"])
```

## Number of triples for ANOVA

```python
def f(t,d):
    return pow(sqrt(1-t*t),d-1)
def P(d):
    return integrate.quad(lambda t: f(t,d)                       , -1, 1)[0]
def mean(d):
    return integrate.quad(lambda t: f(t,d)*(acos(t)-pi/2)**2                , -1, 1)[0]/P(d)
def var(d):
    return integrate.quad(lambda t: f(t,d)*((acos(t)-pi/2)**2-mean(d))**2   , -1, 1)[0]/P(d)

result=[]
for d in range(2,11):
    gap = min( mean(d)-mean(d+1) , mean(d-1)-mean(d) )/2
    stdev = sqrt(var(d))
    n=pow(stdev/gap,2)*10
    result.append([d, gap, stdev, int(n)])
    
pd.DataFrame(result, columns=["dimension", "gap", "standard deviation", "nb of triples"])
```
